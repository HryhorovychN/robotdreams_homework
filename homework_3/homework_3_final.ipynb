```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Set the display parameters
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread('C:/Users/user/anaconda3/sunrise.webp')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Create a blurred version of the original image using Gaussian blurring
blurred = cv2.GaussianBlur(img, (5, 5), sigmaX=1, sigmaY=1)
```

```python
# Compute the difference between the original and the blurred image
diff = img.astype(np.float32) - blurred.astype(np.float32)
```

```python
# Set the sharpening amount
amount = 1.6
```

```python
# Apply the USM filter
sharpened = img.astype(np.float32) + amount * diff
```

```python
sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)
```

```python
plt.imshow(sharpened)
plt.show()
```

```python
# Display the original and sharpened images side by side for comparison
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(sharpened)
axs[1].set_title('Sharpened Image')
axs[1].axis('off')

plt.show()
```
