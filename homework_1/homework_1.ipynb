```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [15, 10]
```

```python
image_path = "C:/Users/user/Documents/GitHub/robotdreams_homework/homework_1/monika_maltipu.jpg"  
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
red, green, blue = cv2.split(img)
```

```python
img1 = cv2.merge([red, green, blue])  # RGB
img2 = cv2.merge([red, blue, green])  # RBG
img3 = cv2.merge([green, red, blue])  # GRB
img4 = cv2.merge([blue, green, red])  # BGR
```

```python
out1 = np.hstack((img1, img2))
out2 = np.hstack((img3, img4))
out = np.vstack((out1, out2))
```

```python
# Plot the collage
plt.imshow(out)
plt.axis(False)
```
